# Data Cleaning

```python
#Extracting the "Date of Publication" column data using Regex
bl_image["Date of Publication"] = bl_image["Date of Publication"].str.extract(r"^(\d{4})",expand = False)
bl_image["Date of Publication"]
```

```python
#Calculating total of one value in column of "Date of Publication"

bl_image["Date of Publication"].value_counts()
```

```python
#Calculating the null values in "Date of Publication" column

bl_image["Date of Publication"].isnull().sum()
```

```python
#Finding the most frequently using value in "Date of Publication" column

bl_image["Date of Publication"].mode()
```

```python
#Filling mode value of "Date of Publication" in place of Nan Values in column

bl_image["Date of Publication"] = bl_image["Date of Publication"].fillna(value = "1897")
```

```python
bl_image.isnull().sum()
```

```python
#Filling Nan Values to "Publisher" column as "Publisher not Known"

bl_image["Publisher"] = bl_image["Publisher"].fillna(value= "Publisher not Known")
```

```python
#Filling Nan Values to "Author" column as "Author not Known"

bl_image["Author"] = bl_image["Author"].fillna(value = "Author not none")
```

```python
#Checking total number of "London" value present in column "Place of Publication"

bl_image["Place of Publication"].str.contains("London").sum()
```

```python
#Checking total number of "Oxford" value present in column "Place of Publication"

bl_image["Place of Publication"].str.contains("Oxford").sum()
```

```python
#Checking all values present in column of "Place of Publication"

bl_image["Place of Publication"].value_counts()
```

```python
#Removing unwanted values present in column "Place of Publication"
bl_image["Place of Publication"] = np.where(bl_image["Place of Publication"].str.contains("London"), 'London',
                                      np.where(bl_image["Place of Publication"].str.contains("Oxford"), 'Oxford',
                                               bl_image["Place of Publication"].str.replace('-', ' ')))
```
